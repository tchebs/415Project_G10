### getDescriptors

This function takes in a list of filenames, corresponding categories, and a SIFT or SURF object. It outputs two arrays. The first array contains the descriptors for each of the images and the second array contains a list of strings that represent the category names for the flowers. 

```python
def getDescriptors(files, categories, detector):

    xtrain=[]
    ytrain=[]

    path = 'flower_classification/training/'
    
    for i, filename in enumerate(files):
        if i % 500 == 0:
            print(i)
            
        img = cv2.cvtColor(cv2.imread(path + filename + ".jpg"), cv2.COLOR_BGR2RGB)

        kp, des = detector.detectAndCompute(img, None)

        label = categories[i]
                
        xtrain.append(des)
        ytrain.append('flower_' + str(label))

    return xtrain, ytrain
```

The runDatagen function calls getDescriptors and converts the output to numpy arrays. It previously also performed a scaling of the data to [-1, 1].

```python
def runDatagen(files_train, cats_train, detector):    
    xtrain, ytrain = getDescriptors(files_train, cats_train, detector)

    print('datagen done')
    # print(xtrain[0], ytrain[0])
    print(len(xtrain))
    print(len(ytrain))
    print(xtrain[0][0])
    print(ytrain[0])
    
    xtrain = np.array(xtrain)
    ytrain = np.array(ytrain)
    
#     print(len(xtrain))
#     print(len(ytrain))
#     print(xtrain[0][0])
#     print(ytrain[0])
    
    return xtrain, ytrain
```

### trainModel

This function takes in the lists of descriptors and categories generated by the runDatagen function. It first transforms the data into a format that can be used to train the SVM. Instead of being a jagged 3d array of descriptors, the descriptors array is converted to a 2d array where each row represents a single descriptor. Additionally, in order to save training runtime, only 5 descriptors from each image's descriptor set are taken. The categories list is modified accordingly to match the new descriptors list. Next, the data is scaled to [-1, 1] in order to further accelerate the training process. The scaling object is returned alongside the model in order to be used later in the testing process, as that data will need to be scaled in the same way in order to work with the model. 

```python
def trainModel(xtrain, ytrain): 
    
    xtrain_mod = []
    ytrain_mod = []
    
    for i, ele in enumerate(xtrain):
        selected = set()
        tmp = []
        while len(selected) < 5:
            idx = random.randint(0, len(ele) - 1)
            if idx not in selected:
                xtrain_mod.append(ele[idx])
                ytrain_mod.append(ytrain[i])
                selected.add(idx)
    
#     print(xtrain_mod)
#     print(ytrain_mod)
    print(Counter(ytrain_mod).most_common(5))
    
    scaling = MinMaxScaler(feature_range=(-1,1)).fit(xtrain_mod)
    xtrain = scaling.transform(xtrain_mod)
    
    svmodel = svm.SVC(kernel = 'rbf', C = 1, gamma = 0.00001)
    
    print('Training start')
    svmodel.fit(xtrain_mod, ytrain_mod)
    print('Training done!')
    return svmodel, scaling
```

### testModel

This function takes in the trained model and the testing data and outputs a list containing the test results. In order to perform the test, the data first needs to be scaled to match the data that the model was trained on. Then, using the model.predict function, the SVM outputs the category that each of the given descriptors were associated with. The descriptor that is matched the most is then determined using the python Counter object and is assumed to be the label associated with the image being tested. 

```python
def testModel(model, features_test, cats_test, scaling):
    accuracy = 0
    
    true_array = []
    pred_array = []
    
    for i, features in enumerate(features_test):
        if i % 100 == 0:
            print(str(i) + ' / ' + str(len(features_test)))

        feat_scaled = scaling.transform(features)
        pred = model.predict(feat_scaled)

        counts = Counter(pred)
        pred_label = counts.most_common(1)[0][0]
        pred_array.append(pred_label)

        actual_label = cats_test[i]
        true_array.append(actual_label)

        if actual_label == pred_label:
            accuracy += 1

        print(str(i) + " - Actual label: " + str(actual_label) + " - Pred label: " + str(pred_label))
        print(counts.most_common())

    accuracy = (accuracy / len(features_test)) * 100
    accuracy
    print("\nAccuracy: %.2f" % accuracy + "%")
    return accuracy, true_array, pred_array
```

# SIFT

Here we define the cross validation function to be used for both SIFT and SURF. This function is explained in the cross validation methodology section of the report. 

```python
sift = cv2.xfeatures2d.SIFT_create()

dataframe = pd.read_csv('flower_classification/training_label.csv')
# dataframe = dataframe.truncate(before = 0, after = 499)
filenames = dataframe['Id'].tolist()
classes = dataframe['Category'].tolist()

print(Counter(classes).most_common())

# xtrain, ytrain = runDatagen(filenames, classes, sift)

def crossValidation(iterations):
    
    accs = []
    trues = []
    preds = []
    
    for i in range(iterations):
        filenames_train_subset = []
        class_train_subset = []
        
        filenames_test_subset = []
        class_test_subset = []
        
        for j in range(1, 103):
            jset = dataframe.loc[dataframe['Category'] == j]
            selected = set()
            while len(selected) < 5:
                idx = random.randint(0, len(jset) - 1)
                if idx not in selected:
                    filenames_train_subset.append(jset.iloc[idx, 0])
                    class_train_subset.append(jset.iloc[idx, 1])
                    selected.add(idx)
            while len(selected) < 6:
                idx = random.randint(0, len(jset) - 1)
                if idx not in selected:
                    filenames_test_subset.append(jset.iloc[idx, 0])
                    class_test_subset.append(jset.iloc[idx, 1])
                    selected.add(idx)
        
        print(Counter(class_train_subset).most_common())
        print(Counter(class_test_subset).most_common())
        
        x_train, y_train = runDatagen(filenames_train_subset, class_train_subset, sift)
        x_test, y_test = runDatagen(filenames_test_subset, class_test_subset, sift)
        
#         x_train, x_test, y_train, y_test = train_test_split(xtrain, ytrain, test_size = 0.2, random_state = i)
#         kf = KFold(n_splits = 5)
#         train_index, test_index = kf.split(xtrain, ytrain)
#         print(train_index)
#         print(test_index)
        
#         print(x_train.shape)
#         print(x_train[0].shape)
#         print(x_test.shape)
#         print(y_train.shape)
#         print(y_test.shape)
#         print(x_train)
#         print(y_train)
        model, scaler = trainModel(x_train, y_train)
        acc, true, pred = testModel(model, x_test, y_test, scaler)
        accs.append(acc)
        trues.append(true)
        preds.append(pred)
        
    print(accs)
    return accs, trues, preds

accs, trues, preds = crossValidation(2)
```

### compute_metrics

This function takes in the results of testModel and computes the metrics required for the report. Specifically, we compute accuracy, standard deviation of accuracy, average recall, and average precision. We also compute and display the confusion matrix to be shown in the report. 

```python
# print(trues)
# print(preds)
def compute_metrics(trues, preds):
    accs = [sklearn.metrics.accuracy_score(trues[i], preds[i]) for i in range(len(trues))]
    recalls = [sklearn.metrics.recall_score(trues[i], preds[i], average = None) for i in range(len(trues))]
    precs = [sklearn.metrics.precision_score(trues[i], preds[i], average = None) for i in range(len(trues))]
    
    conf_mat = sklearn.metrics.confusion_matrix(trues[0], preds[0])
    
    print(np.mean(accs))
    print(np.std(accs))
    print(recalls)
    print(np.mean(recalls))
    print(precs)
    print(np.mean(precs))
    
    plt.figure(figsize = (40, 40))
    disp = sklearn.metrics.ConfusionMatrixDisplay(confusion_matrix = conf_mat, display_labels = trues[0])
    disp.plot(xticks_rotation = 'vertical', ax = plt.gca())
    plt.show()
    
compute_metrics(trues, preds)
```
